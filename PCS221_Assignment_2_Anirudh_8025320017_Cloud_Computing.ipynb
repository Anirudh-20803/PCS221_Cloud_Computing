*1. Create a data file by using Python random module with different random generator parameters. The data file must have 20000 X 10000 (rows X columns) in the data file. This data file to be used as database for realization of math and matrix operations in cloud computing lab assignments.*

```python
import random
ROWS = 20000
COLS = 10000
with open("cloud_lab_data.csv", "w") as file:
    for i in range(ROWS):
        row_data = []
        for j in range(COLS):
            value = random.random()
            row_data.append(str(value))
        file.write(",".join(row_data) + "\n")
        if i % 100 == 0:
            print(f"{i} are completed")
print("The data file is created successfully")
```

*2. Write a program in Python to assign subsequent rows to matrix first row elements.*

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Matrix input by user")
for row in matrix:
    print(row)
for i in range(1, len(matrix)):
    matrix[0] = matrix[i][:]
    print("\nAfter assigning row", i+1, "to first row:")
    for row in matrix:
        print(row)
```

*3. Write a program in Python to group similar elements into matrix.*

```python
elements = [1, 2, 1, 3, 2, 1, 4, 3]
grouped = {}
for item in elements:
    if item in grouped:
        grouped[item].append(item)
    else:
        grouped[item] = [item]
matrix = list(grouped.values())
print("Grouped Matrix:")
for row in matrix:
    print(row)
```

*4. Write a program in Python to add and subtract matrices in python.*

```python
A = [
    [1, 2, 3],
    [4, 5, 6]
]
B = [
    [7, 8, 9],
    [10, 11, 12]
]
rows = len(A)
cols = len(A[0])
addition = []
subtraction = []
for i in range(rows):
    add_row = []
    sub_row = []
    for j in range(cols):
        add_value = A[i][j] + B[i][j]
        sub_value = A[i][j] - B[i][j]
        add_row.append(add_value)
        sub_row.append(sub_value)
    addition.append(add_row)
    subtraction.append(sub_row)
print("Addition of matrices:")
for row in addition:
    print(row)
print("\nSubtraction of matrices:")
for row in subtraction:
    print(row)
```

*5. Write a program in Python to create an n x n square matrix, where all the sub-matrix has the sum of opposite corner elements as even.*

```python
n = int(input("Enter value of n: "))
matrix = []
for i in range(n):
    row = []
    for j in range(n):
        value = (i + j) % 2
        row.append(value)
    matrix.append(row)
print("\nGenerated Matrix:")
for row in matrix:
    print(row)
```

*6. Write a program in Python to perform row-wise element addition in tuple matrix.*

```python
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))
matrix = []
print("Enter elements row-wise:")
for i in range(rows):
    row = []
    for j in range(cols):
        value = int(input(f"Element [{i}][{j}]: "))
        row.append(value)
    matrix.append(tuple(row))
matrix = tuple(matrix)
print("\nRow-wise element addition:")
for i in range(len(matrix)):
    row_sum = 0
    for j in range(len(matrix[i])):
        row_sum += matrix[i][j]
    print("Sum of row", i + 1, "=", row_sum)
```
